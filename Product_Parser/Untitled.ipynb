```python
from selenium import webdriver
```

```python
driver = webdriver.Chrome()
```

```python
url = input()
cn = input()
cns = input()
```

```python
driver.get(url)
```

```python
imgs = driver.find_elements_by_xpath('//div[@class="a-section a-spacing-none aok-relative"]/span/a/span/div/img')
```

```python
imgs = [img.get_attribute('src') for img in imgs]
```

```python
imgs
```

```python
for img in imgs:
    print(img)
    print()
```

```python
len(imgs)
```

```python
for i in range(len(imgs)):
    try:
        urllib.request.urlretrieve(imgs[i], cns+str(i)+".jpg")
        print(i)
    except(e):
        print(e)
        continue
```

```python
names = driver.find_elements_by_xpath('//div[@class="a-section a-spacing-none aok-relative"]/span/a/div')
```

```python
names = [name.text for name in names]
```

```python
names
```

```python
prices = driver.find_elements_by_xpath('//div[@class="a-section a-spacing-none aok-relative"]/span/div[@class="a-row"]/a/span/span')
```

```python
prices = [price.text for price in prices]
```

```python
prices = [price.split(' ')[1] for price in prices]
```

```python
prices = [''.join(price.split(',')) for price in prices]
```

```python
prices = [int(price.split('.')[0]) for price in prices]
```

```python
len(prices)
```

```python
df = pd.DataFrame(columns = ['product_name','price','image','category'])
```

```python
for i in range(35):
    df.loc[i,'product_name'] = names[i]
    df.loc[i,'price'] = prices[i]
    df.loc[i,'image'] = 'shop/images/'+cns+str(i)+'.jpg'
    df.loc[i,'category'] = cn 
    df.loc[i,'subcategory'] = cns+str(i)
```

```python
df.to_csv(cns+'prods.csv',index=False)
```

```python

for i in range(len(df)):
    flag=0
    df.loc[i,'product_desc'] = df.loc[i,'product_name']
    name=df.loc[i,'product_name']
    if(',' in name):
        name = name.split(',')[0]
        flag=1
    if('-' in name):
        name = name.split('-')[0]
        flag=1
    if('(' in name):
        name = name.split('(')[0]
        flag=1
    if(flag==0 or len(name)>5):
        name = ' '.join(name.split(' ')[:5])
    df.loc[i,'product_name'] = name
```
